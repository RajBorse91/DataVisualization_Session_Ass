### Bivariate **Analysis**

```python
x1 = day.Active
y1 = day.Recovered

x2 = day.Active
y2 = day.Deaths

plt.figure(figsize=(12,10))
plt.scatter(x1, y1, c="red", s=20, label="Recoverd Cases")
plt.scatter(x2, y2, c="green", s=20, label="Death's")
plt.grid()
plt.title("RECOVERD & DEATHS")
plt.xlabel("$ACTIVE$")
plt.ylabel("$RECOVER/DEATHS$")
# For enabling the labeling
plt.legend(loc="best")
plt.show()
```



*   After the cases are increased the reovery rate also has been increased but the death ration is maintained i.e; death rate is not much affected. 
*   As the cases are increased the recovery rate has been also incresed
